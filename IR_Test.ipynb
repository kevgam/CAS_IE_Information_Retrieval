```python
def find_invalid_lyrics_indices(df):
    """
    Function to find the indices of rows in the DataFrame where the 'lyrics' column
    has no entry (NaN or empty) or contains an integer value.

    Args:
        df (pd.DataFrame): DataFrame to check.

    Returns:
        list: A list of tuples with the index and reason ('empty' or 'integer').
    """
    invalid_indices = []

    for index, value in df['lyrics'].items():
        if pd.isna(value) or str(value).strip() == '':
            invalid_indices.append((index, 'empty'))
        elif isinstance(value, int):
            invalid_indices.append((index, 'integer'))

    return invalid_indices
```

```python
import pandas as pd
from langdetect import detect
from datetime import datetime

# Filepath to the CSV
filepath = '/content/drive/MyDrive/ie_information_retrieval_dataset/songs_with_attributes_and_lyrics.csv'

# Load the CSV into a DataFrame
dfs = pd.read_csv(filepath)

# Define a function to detect language, handling exceptions
def detect_language(text):
    try:
        return detect(text)
    except:
        return 'unknown'  # Return 'unknown' if detection fails

# Process rows in batches and log progress
def process_lyrics_with_logging(df, batch_size=5000):
    """
    Process the 'lyrics' column to detect language, logging progress every batch_size rows.
    Saves the processed DataFrame as an XLSX file in Google Drive after completion.

    Args:
        df (pd.DataFrame): DataFrame containing the data.
        batch_size (int): Number of rows to process before logging progress.

    Returns:
        pd.DataFrame: DataFrame with 'lyrics_language' column updated.
    """
    start_time = datetime.now()
    print(f"Processing started at: {start_time}")

    # Iterate over rows
    for index, row in df.iterrows():
        # Apply language detection
        df.at[index, 'lyrics_language'] = detect_language(row['lyrics'])

        # Log progress every batch_size rows
        if (index + 1) % batch_size == 0:
            current_time = datetime.now()
            print(f"Processed {index + 1} rows at {current_time}")

    # Save the final DataFrame as an XLSX file
    output_path = '/content/drive/MyDrive/ie_information_retrieval_dataset/processed_songs_with_lyrics.xlsx'
    df.to_excel(output_path, index=False)

    end_time = datetime.now()
    print(f"Processing completed at: {end_time}")
    print(f"DataFrame saved to: {output_path}")
    return df

# Initialize the 'lyrics_language' column if not already present
if 'lyrics_language' not in dfs.columns:
    dfs['lyrics_language'] = None

# Process the DataFrame and save it
dfs = process_lyrics_with_logging(dfs)
```

```python
import pandas as pd
from langdetect import detect
from concurrent.futures import ThreadPoolExecutor, as_completed
from datetime import datetime

# Define a function to detect language, handling exceptions
def detect_language(text):
    try:
        return detect(text)
    except:
        return 'unknown'  # Return 'unknown' if detection fails

def process_lyrics_in_parallel(df, batch_size=5000, num_threads=8):
    """
    Process the 'lyrics' column to detect language in parallel batches, logging progress every batch_size rows.

    Args:
        df (pd.DataFrame): DataFrame containing the data.
        batch_size (int): Number of rows to process before logging progress.
        num_threads (int): Number of threads to use for parallel processing.

    Returns:
        pd.DataFrame: DataFrame with 'lyrics_language' column updated.
    """
    start_time = datetime.now()
    print(f"Processing started at: {start_time}")

    # Split the DataFrame into batches
    batches = [df.iloc[i:i + batch_size] for i in range(0, len(df), batch_size)]

    # Process each batch in parallel
    results = []
    with ThreadPoolExecutor(max_workers=num_threads) as executor:
        future_to_batch = {executor.submit(process_batch, batch): batch for batch in batches}

        for future in as_completed(future_to_batch):
            results.append(future.result())
            print(f"Processed {len(results) * batch_size} rows at {datetime.now()}")

    # Combine results into a single DataFrame
    processed_df = pd.concat(results)

    # Save the final DataFrame as an XLSX file
    output_path = '/content/drive/MyDrive/processed_songs_with_lyrics.xlsx'
    processed_df.to_excel(output_path, index=False)

    end_time = datetime.now()
    print(f"Processing completed at: {end_time}")
    print(f"DataFrame saved to: {output_path}")

    return processed_df

def process_batch(batch):
    """
    Process a single batch of rows to detect language.

    Args:
        batch (pd.DataFrame): A subset of the DataFrame.

    Returns:
        pd.DataFrame: The processed batch with 'lyrics_language' updated.
    """
    batch['lyrics_language'] = batch['lyrics'].apply(detect_language)
    return batch

# Filepath to the CSV
filepath = '/content/drive/MyDrive/ie_information_retrieval_dataset/songs_with_attributes_and_lyrics.csv'

# Load the CSV into a DataFrame
dfs = pd.read_csv(filepath)

# Initialize the 'lyrics_language' column if not already present
if 'lyrics_language' not in dfs.columns:
    dfs['lyrics_language'] = None

# Process the DataFrame in parallel
dfs = process_lyrics_in_parallel(dfs)
```
